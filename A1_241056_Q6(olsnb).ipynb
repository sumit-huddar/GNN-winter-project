```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('/content/housing_prices - housing_prices.csv')
x_data = df.iloc[:, 0].values
y_data = df.iloc[:, 1].values
```

```python
X_matrix = np.c_[np.ones(x_data.shape[0]), x_data]
XtX = X_matrix.T @ X_matrix
XtX_inv = np.linalg.inv(XtX)
Xty = X_matrix.T @ y_data
theta = XtX_inv @ Xty
b, m = theta[0], theta[1]
```

```python
prediction_sqft = 2500
predicted_price = m * prediction_sqft + b
print(f"OLS Predicted price for {prediction_sqft} sq ft: ${predicted_price:.2f}")

# --- 4. Visualization ---
plt.figure(figsize=(10, 6))
plt.scatter(x_data, y_data, color='blue', label='Actual Data')

# Create a line for plotting
x_line = np.linspace(x_data.min(), x_data.max(), 100)
y_line = m * x_line + b

plt.plot(x_line, y_line, color='red', label=f'OLS Line: y={m:.2f}x + {b:.2f}')
plt.xlabel('Square Footage')
plt.ylabel('Price')
plt.title('Housing Prices: OLS Method')
plt.legend()
plt.grid(True)
plt.show()
```
